```python
class Employee:
    raise_rate = 1.1
    
    def __init__(self,name,surname,salary):
        self.name = name
        self.surname = surname
        self.salary = salary
        self.email = name + surname + "@gmail.com"
    
    def show_info(self):
        return f"Name: {self.name}, Surname: {self.surname}, Salary: {self.salary}, E-mail: {self.email}"

employee1 = Employee("Ali","Tandoğan",25000)
employee2 = Employee("Olcay","Çakır",37000)

class Developer(Employee): ## Employee class'ının özelliklerini miras alacağı için içine Employee yazmalıyız.
    ##pass ##Eğer class içini sonradan doldurucaksak pass komutu ile bırakabiliriz.
    
    def __init__(self,name,surname,salary,softwarelanguage):
        ##self.name = name
        ##self.surname = surname
        ##self.salary = salary
        ##self.email = name + surname + "@gmail.com"
        super().__init__(name,surname,salary) ##Aslında yukarıdaki gibi yapmak yerine bu şekilde de yapabiliriz.
        self.softwarelanguage = softwarelanguage
##Eğer init fonksiyonunu tanımlamamaış olsaydık yukarıdan alıcaktı. Fakat ekstra özellik eklemek istediğimiz için init fonk
##tanımladık. Yukarıdaki sırayla içini doldurmak daha mantıklı ve kolay oluyor.
    
    raise_rate = 1.2
    
    def show_info(self): ##Override etmiş olduk.
        return f"Name: {self.name}, Surname: {self.surname}, Salary: {self.salary}, E-mail: {self.email}, Language = {self.softwarelanguage}"
    
    def which_language(self):
        return f"Known language is: {self.softwarelanguage}"

class Manager(Employee):
    
    def __init__(self,name,surname,salary,employees = None):
        super().__init__(name,surname,salary)
        if employees == None:
            self.employees = []
        else:
            self.employees = employees
    
    def add_employee(self,employee):
        if employee not in self.employees:
            self.employees.append(employee)
            
    def delete_employee(self,employee):
        if employee in self.employees:
            self.employees.remove(employee)
            
            
            
    def show_employee(self):
        for employee in self.employees:
            print(employee.show_info())
            
            
            
developer1 = Developer("Aydın","Karabulut",42000,"C#")
developer2 = Developer("Ergin","Ataman",120000,"Java")

print(developer2.email)
print(employee2.raise_rate)
print(developer1.raise_rate) ##Eğer Developer class'ının içinde raise_raet olmasa Employee class'ındakini yazdırırdı.

print(employee2.show_info())
print(developer1.show_info())
print(developer2.show_info())
print(developer2.which_language())

manager1 = Manager("Metin","Ali",250000)

manager1.add_employee(employee1)
manager1.add_employee(developer1)
manager1.show_employee()
print("*********")
manager1.delete_employee(employee1)
manager1.show_employee()
print("*********")
manager2 = Manager("Yıldız","Yaren",125000,[developer1,developer2,employee2])
manager2.show_employee()
print("*********")
print(isinstance(manager2,Employee)) ## Manager2 Employee sınıfının bir instance'dır.
print(isinstance(developer1,Manager)) ##Developer1 Manager sınıfının bir instance'ı değildir
print("*********")
print(issubclass(Employee,Developer)) ##Employee class'ı Developer classından türedi mi ? False
print(issubclass(Developer,Employee)) ##Ama Developer class'ı Employee class'ından türemiştir.
```
