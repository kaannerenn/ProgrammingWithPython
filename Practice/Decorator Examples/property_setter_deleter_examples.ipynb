```python
class Person:
    def __init__(self,name,surname,age):
        self.name = name
        self.surname = surname
        self._age = age
    
    @property
    def age(self):
        return f"{self._age}"
    @age.setter
    def age(self,age):
        if age <= 0:
            print("Age must be greater than 0")
        else:
            self._age = age
    @age.deleter
    def age(self):
        print("Age can not be deleted.")
        
    def __str__(self): 
        return f"Name: {self.name}, Surname: {self.surname}, Age: {self._age}"

person1 = Person("Kaan", "Özyılmaz", 25)

person1.age = -5  
print(person1.age)

person1.age = 30
print(person1.age)

del person1.age
print(person1.age)

print(person1)
```

```python
class BankAccount:
    def __init__(self, account_holder, pin, balance=0):
        self.account_holder = account_holder
        self.balance = balance
        self._pin = pin   ##private
    
    @property
    def pin(self):
        return self._pin  
    
    @pin.setter
    def pin(self, newpin):
        if len(str(newpin)) == 4:  
            self._pin = newpin
        else:
            print("New pin must be 4 characters.")  
    
    @pin.deleter
    def pin(self):
        
        print("PIN deleted! Please set a new PIN.")
        new_pin = int(input("Enter a new 4-digit PIN: "))
        self.pin = new_pin  


bankaccount1 = BankAccount("Yılmaz Vural", 1234)
bankaccount1.pin = 34567
print(bankaccount1.pin)  
print("****")
bankaccount1.pin = 1289
print(bankaccount1.pin)
print("****")
del bankaccount1.pin  
print(bankaccount1.pin)  
```

```python
class Book:
    def __init__(self,title,author,price,isbn):
        self.title = title
        self.author = author
        self.price = price
        self._isbn = isbn
        
    def __str__(self):
        
        return f"title = {self.title}, author = {self.author}, price = {self.price}, isbn = {self._isbn}\n**"
    
    @property
    def isbn(self):
        return self._isbn
    
    @isbn.setter
    def isbn(self,new_isbn):
        if len(str(new_isbn)) == 13:
            self._isbn = new_isbn
        else:
            print("ISBN must be 13 digits")
    
    @isbn.deleter
    def isbn(self):
        self._isbn = None
        print("ISBN number has been reset")

book1 = Book("Araba Sevdası","Namık Kemal",130,1234567891011)
print(book1)
print(book1.isbn)
book1.isbn = 132456
```

```python
class Product:
    def __init__(self,product_name,category,sku,price = 0):
        self.product_name = product_name
        self.category = category
        self._sku = sku
        self.price = price
        
    @property
    def sku(self):
        return self._sku
    
    @sku.setter
    def sku(self,new_sku):
        if len(str(new_sku)) == 10:
            self._sku = new_sku
        else:
            print("SKU number must be 10 digits")
    
    @sku.deleter
    def sku(self):
        self._sku = None
        print("SKU number has been reset")
        
    def __str__(self):
        return f"{self.product_name}, {self.category}, {self.price}, {self.sku}"
        
        
product1 = Product("Laptop", "Electronics", "1234567890", 1500)
print(product1)

product1.sku = "0987654321"
print(product1)

del product1.sku
print(product1)
```

```python
class Product:
    def __init__(self, name, category, price, stock, discount=0):
        self.name = name
        self.category = category
        self._price = price
        self._stock = stock
        self._discount = discount

    @property
    def price(self):
        # İndirim uygulanmış fiyatı döndürür
        if self._discount > 0:
            return round(self._price * (1 - self._discount / 100), 2)
        return self._price

    @property
    def stock(self):
        return self._stock

    @stock.setter
    def stock(self, value):
        if value < 0:
            print("Stock cannot be negative!")
        else:
            self._stock = value
            if self._stock == 0:
                print("Ürün tükenmiştir")

    @property
    def discount(self):
        return self._discount

    @discount.setter
    def discount(self, value):
        if 0 <= value <= 100:
            self._discount = value
        else:
            print("Discount must be between 0 and 100")

    @discount.deleter
    def discount(self):
        self._discount = 0
        print(f"Discount has been reset. Price is back to original: {self._price}")

    def __str__(self):
        return f"Product: {self.name}, Category: {self.category}, Price: {self.price}, Stock: {self.stock}, Discount: {self._discount}%"

# Örnek kullanım:
product1 = Product("Laptop", "Electronics", 1000, 10)

print(product1)  # Başlangıç durumu

product1.discount = 20  
print(product1)  

product1.stock = -5 
product1.stock = 0  

del product1.discount  
print(product1) 
```
