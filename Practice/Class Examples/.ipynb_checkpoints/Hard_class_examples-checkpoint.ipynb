```python
class Book:
    
    book_list = []
    
    def __init__(self,book_name,author,page_number,stock_quantity):
        self.book_name = book_name
        self.author = author
        self.page_number = page_number
        self.stock_quantity = stock_quantity
        Book.book_list.append(self)
    
    @classmethod
    def add_books(cls, *books):
        for book in books:
            if isinstance(book, cls):  # Sadece Book veya türevlerini ekle
                cls.book_list.append(book)
            else:
                print(f"{book} is not a valid Book instance!")

    @classmethod
    def get_book_list(cls):
        return "\n".join(str(book) for book in cls.book_list)
        
    def __str__(self):
        return f"{self.book_name} -- {self.author} -- {self.page_number} -- {self.stock_quantity}"
    
    def update_stock(self,new_stock):
        self.stock_quantity = new_stock
    
    def book_borrow(self,quantity = 1):
        
            if self.stock_quantity >= quantity:
                self.stock_quantity -= quantity
            else:
                print("Out of stock.")
            
    def book_return(self, quantity = 1):
            self.stock_quantity += quantity
            
    def __eq__(self,other):
        if self.book_name.lower() == other.book_name.lower() and self.author.lower() == other.author.lower():
            return "Bu kitaplar aynı kitaplardır."
        return "Bunlar kitaplar aynı değildir."
    
    def __lt__(self,other):
        if self.page_number < other.page_number:
            return f"{self.book_name} kitabının sayfa sayısı {other.book_name} kitabından azdır."
        return f"{self.book_name} kitabının sayfa sayısı {other.book_name} kitabından fazladır."
    
    def __gt__(self,other):
        if self.page_number > other.page_number:
            return f"{self.book_name} kitabının sayfa sayısı {other.book_name} kitabından fazladır."
        return f"{self.book_name} kitabının sayfa sayısı {other.book_name} kitabından azdır."
    
    
    
class Novel(Book):
    def __init__(self,book_name,author,page_number,stock_quantity,novel_type):
        super().__init__(book_name,author,page_number,stock_quantity)
        self.novel_type = novel_type
        
    def __str__(self):
        return f"{self.book_name} -- {self.author} -- {self.page_number} -- {self.stock_quantity} -- {self.novel_type}"

book1 = Book("Peter pan","Kaan eren",23,2)
print(str(book1))
book1.update_stock(4)
print(str(book1))
book1.book_borrow()
print(str(book1))
book1.book_return()
print(str(book1))
novel1 = Novel("Don kişot","Buse",56,20,"Adventure")
print(str(novel1))
novel1.update_stock(22)
print(str(novel1))
novel1.book_borrow(20)
print(str(novel1))
novel1.book_return(10)
print(str(novel1))
book2 = Book("peter pan","kaan EREN",23,4)
print(book1 == book2)
print(book1 == novel1)
print(book1 < novel1)
print(novel1 > book2)
print(Book.get_book_list())
book4 = Book("Araba sevdası", "", 102, 4)
book5 = Novel("Sherlock Holmes", "Arthur Conan Doyle", 320, 15, "Detective")
Book.add_books(book4, book5)  
print(Book.get_book_list())
```

```python
class Book:
    book_list = []
    
    def __init__(self, book_name, author, page_number, book_type, cost, quantity):
        self.book_name = book_name
        self.author = author
        self.page_number = page_number
        self.book_type = book_type
        self.cost = cost
        self.quantity = quantity
        Book.book_list.append(self)
        
    def __str__(self):
        return f"Kitap Adı: {self.book_name}, Yazar: {self.author}, Sayfa Sayısı: {self.page_number}, Tür: {self.book_type}, Fiyat: {self.cost} TL, Adet: {self.quantity}"
        
    @classmethod
    def show_book_list(cls):
        return "\n".join(str(book) for book in cls.book_list)
    
    def __lt__(self, other):
        if self.book_type.lower() == other.book_type.lower():
            if self.cost < other.cost:
                return f"{self.book_name} kitabı {other.book_name} kitabından {other.cost - self.cost} TL daha ucuzdur."
            elif self.cost == other.cost:
                return f"{self.book_name} kitabı {other.book_name} kitabıyla aynı fiyattadır."
            else:
                return f"{other.book_name} kitabı {self.book_name} kitabından {self.cost - other.cost} TL daha ucuzdur."
        else:
            return "Sadece aynı tür kitapların fiyatı karşılaştırılabilir."
            
    def __gt__(self, other):
        if self.book_type.lower() == other.book_type.lower():
            if self.cost > other.cost:
                return f"{self.book_name} kitabı {other.book_name} kitabından {self.cost - other.cost} TL daha pahalıdır."
            elif self.cost == other.cost:
                return f"{self.book_name} kitabı {other.book_name} kitabıyla aynı fiyattadır."
            else:
                return f"{other.book_name} kitabı {self.book_name} kitabından {other.cost - self.cost} TL daha pahalıdır."
        else:
            return "Sadece aynı tür kitapların fiyatı karşılaştırılabilir."
            
class Publisher(Book):
    discounted_list = []
    
    def __init__(self, book_name, author, page_number, book_type, cost, quantity, publisher):
        super().__init__(book_name, author, page_number, book_type, cost, quantity)
        self.publisher = publisher
        
    @classmethod
    def discounted_books(cls):
        cls.discounted_list = [book for book in Book.book_list if book.quantity <= 5]
        return "\n".join(str(book) for book in cls.discounted_list)

class Sale(Publisher):
    money = 0
    
    @classmethod
    def sell(cls, book, quantity):
        if book.quantity >= quantity:
            book.quantity -= quantity
            cls.money += book.cost * quantity
        else:
            print(f"Elimizde {book.book_name} kitabı {quantity} adet mevcut değil.")
            
# Kitaplar
book1 = Book("Book 1", "Author 1", 200, "Fiction", 50, 10)
book2 = Book("Book 2", "Author 2", 150, "Non-Fiction", 40, 3)
book3 = Book("Book 3", "Author 3", 300, "Fiction", 60, 6)

# Publisher
publisher = Publisher("Book 1", "Author 1", 200, "Fiction", 50, 10, "Publisher 1")

# Karşılaştırmalar
print(book1.__lt__(book2))
print(book1.__gt__(book2))

# İndirimli kitaplar
print(Publisher.discounted_books())

# Satış işlemi
Sale.sell(book1, 2)
print(f"Satıştan elde edilen gelir: {Sale.money} TL")
```
