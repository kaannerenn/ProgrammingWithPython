```python
from datetime import datetime
class Market:
    total_product = 0
    product_list = []
    
    def __init__ (self,product_name,product_type,cost):
        self.product_name = product_name
        self.product_type = product_type
        self.cost = cost
        Market.total_product += 1
        Market.product_list.append(self)
    
    @classmethod
    def show_all_products(cls):
        for product in cls.product_list:
            print(f"{product.product_name} - {product.product_type} - {product.cost} TL")
    
    @classmethod
    def show_product_number(cls):
        print(f"Toplam ürün sayısı = {cls.total_product}")
    
    @classmethod
    def search_product(cls, product_name):
        for product in cls.product_list:
            if product.product_name.lower() == product_name.lower():
                return f"Aradığınız {product_name} ürünü mevcuttur. Fiyatı {product.cost} TL"
        return f"{product_name} markette yok"
    
    @staticmethod
    def is_market_open():
        current_time = datetime.now()
        formatted_time = current_time.strftime("%H:%M")
        if 9 <= current_time.hour < 22:
            print(f"Time is = {formatted_time} Market is open")
        else:
            print(f"Time is = {formatted_time} Market is close")
    
    def update_cost(self,new_price):
        self.cost = new_price
        print(f"Fiyat güncellenmiştir yeni fiyat = {self.cost}")
        
    def show_cost(self):
        print(f"Ürünün fiyatı = {self.cost}")
    
    @classmethod
    def delete_product(cls,product_name):
        for product in cls.product_list:
            if product_name.lower() == product.product_name.lower():
                cls.product_list.remove(product)
                cls.total_product -= 1
                print(f"{product_name} ürünü silindi")
    @classmethod
    def search_by_type(cls, product_type):
        found_products = [product for product in cls.product_list if product.product_type.lower() == product_type.lower()]
        if found_products:
            for product in found_products:
                print(f"{product.product_name} - {product.product_type} - {product.cost} TL")
        else:
            print(f"{product_type} türünde ürün bulunamadı.")   
    
    @classmethod 
    def search_by_price(cls, min_price, max_price):
        found_products = [product for product in cls.product_list if min_price <= product.cost <= max_price]
        if found_products:
            for product in found_products:
                print(f"{product.product_name} - {product.product_type} - {product.cost} TL")
        else:
            print(f"{min_price} TL ile {max_price} TL arasında ürün bulunamadı.")   

        
product1 = Market("Halley","Snack",38)
product2 = Market("Cif","Cleaning Material",25)
product3 = Market("Hazelnut","Nuts",120)
Market.show_product_number()
Market.is_market_open()
Market.show_all_products()
print(Market.search_product("Halley"))
print(Market.search_product("Abc"))
Market.update_cost(product1,40)
product1.show_cost()
Market.delete_product("Abc")
Market.search_by_type("Snack")
Market.search_by_price(15,30)
```

```python
class Employee:

    
    def __init__(self,name,employee_type,age,salary):
        self.name = name
        self.employee_type = employee_type
        self.age = age
        self.salary = salary
    
    def show_info(self):
        return f"{self.name}, {self.employee_type}, {self.salary}"

class Intern(Employee):
    
    def __init__(self,name,employee_type,age):
        self.name = name
        self.employee_type = employee_type
        self.age = age
        self.salary = 0
        
    def show_info(self):
        return f"{self.name}, {self.employee_type}, No Salary"
    
class Manager(Employee):
    
    def __init__(self,name,employee_type,age,salary,bonus):
        super().__init__(name,employee_type,age,salary)
        self.salary += bonus
        
        
employee1 = Manager("Hasan","Manager",33,25000,3000)
employee2 = Intern("İlayda","İntern",19)
employee3 = Employee("Mert","General Employee",28,23000)

print(employee1.show_info())
print(employee3.show_info())
print(employee2.show_info())
```

```python
class Employee:
    def __init__(self, name, age, base_salary):
        self.name = name
        self.age = age
        self.base_salary = base_salary

    def calculate_salary(self):
        return self.base_salary

class FullTimeEmployee(Employee):
    def __init__(self, name, age, base_salary, bonus):
        # Employee sınıfını çağırıyoruz
        super().__init__(name, age, base_salary)
        self.bonus = bonus

    def calculate_salary(self):
        # Full-time çalışan maaşı base_salary + bonus
        return super().calculate_salary() + self.bonus

class PartTimeEmployee(Employee):
    def __init__(self, name, age, base_salary, hours_worked):
        super().__init__(name, age, base_salary)
        self.hours_worked = hours_worked

    def calculate_salary(self):
        # Part-time çalışan maaşı base_salary + saat başı ücret
        hourly_rate = 20  # Örnek olarak saat başı ücret
        return super().calculate_salary() + (self.hours_worked * hourly_rate)

# Çalışanları tanımlayalım
full_time_employee = FullTimeEmployee("Hasan", 30, 4000, 500)
part_time_employee = PartTimeEmployee("Mert", 22, 1500, 20)

# Çalışanların maaşlarını hesaplayalım
print(f"{full_time_employee.name} full-time maaşı: {full_time_employee.calculate_salary()} TL")
print(f"{part_time_employee.name} part-time maaşı: {part_time_employee.calculate_salary()} TL")
```

```python
class Vehicle:
    def __init__(self,brand,model):
        self.brand = brand
        self.model = model

class Car(Vehicle):
    def __init__(self,brand,model,mass):
        super().__init__(brand,model)
        self.mass = mass
        
    def calculate_speed(self):
        return f"Your speed limit is {(self.model / self.mass) * 15}"
    
class Truck(Vehicle):
    def __init__(self,brand,model,mass):
        super().__init__(brand,model)
        self.mass = mass
    
    def calculate_speed(self):
        return f"Your speed limit is {self.model / (self.mass * 2) * 13}"

vehicle1 = Car("Toyota", 2022, 500)
vehicle2 = Truck("Scania",2020,1000)
print(vehicle1.calculate_speed())
print(vehicle2.calculate_speed())
```
